```python
import pandas as pd
import numpy as np
import sklearn.metrics as m
```

```python
wind_csv_path = "../data/cleaned/bandarAbas_multi_var_wind_3hourly_cleaned.csv"
wind_hourly = pd.read_csv(wind_csv_path)
```

```python
wind_hourly.head()
```

```python
train = wind_hourly.iloc[:-720]
test = wind_hourly.iloc[-720:]
```

```python
pred = test.wind_speed.shift(1).dropna()
obs = test.wind_speed[1:]
```

```python
mae = m.mean_absolute_error(obs, pred)
rmse = m.mean_squared_error(obs, pred)
mape = np.mean(np.abs((obs - pred)/obs))*100
```

```python
 #--- OUTPUT THE RESULTS ---
print("--- Baseline Forecast Error Metrics ---")
print(f"Total Observations: {len(obs):,}")
print("-" * 35)
print(f"Mean Absolute Error (MAE):         {mae:.3f}")
print(f"Root Mean Squared Error (RMSE):    {rmse:.3f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
print("-" * 35)
```
